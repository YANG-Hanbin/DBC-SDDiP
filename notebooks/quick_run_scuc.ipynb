# SCUC DBC/iDBC Notebook

Use this notebook to run minimal SCUC experiments with either `DBC` or `iDBC`.


```julia
using Pkg

function find_repo_root(start::AbstractString = pwd())
    dir = abspath(start)
    while true
        has_project = isfile(joinpath(dir, "Project.toml"))
        has_entry = isfile(joinpath(dir, "experiments", "scuc", "run_experiment.jl"))
        if has_project && has_entry
            return dir
        end
        parent = dirname(dir)
        parent == dir && error("Could not locate repository root from $(start).")
        dir = parent
    end
end

repo_root = find_repo_root()
cd(repo_root)
Pkg.activate(repo_root)
println("repo_root = ", repo_root)
```

## Edit Parameters

`cut_mode` can be `:DBC` or `:iDBC`.


```julia
run_script = joinpath("experiments", "scuc", "run_experiment.jl")
config_path = joinpath("notebooks", "tmp_scuc_notebook_config.jl")

cut_mode = :DBC                     # :DBC or :iDBC
enable_copy_branching = (cut_mode == :iDBC)
enable_surrogate_copy_split = true
idbc_warm_pass = false

periods = 6
realizations = 5

num_workers = 1
dry_run = false
max_iter = 5
num_scenarios = 5
num_backward_scenarios = 1
num_partition_scenarios = 1
terminate_time = 300.0
time_limit = 20.0
experiment_tag = "nb_scuc_$(cut_mode)"

config_text = """
EXPERIMENT_CONFIG = ScucExperimentConfig(
    num_workers = $(num_workers),
    dry_run = $(dry_run),
    static = ScucStaticConfig(
        max_iter = $(max_iter),
        num_scenarios = $(num_scenarios),
        num_backward_scenarios = $(num_backward_scenarios),
        num_partition_scenarios = $(num_partition_scenarios),
        terminate_time = $(terminate_time),
        time_limit = $(time_limit),
        enable_copy_branching = $(enable_copy_branching),
        enable_surrogate_copy_split = $(enable_surrogate_copy_split),
        copy_split_strategy = :surrogate_delta,
        idbc_warm_pass = $(idbc_warm_pass),
        experiment_tag = $(repr(experiment_tag)),
        legacy_logger_paths = false,
    ),
    sweep = ScucSweepConfig(
        algorithms = Symbol[:SDDiP],
        cuts = Symbol[$(repr(cut_mode))],
        periods = Int[$(periods)],
        realizations = Int[$(realizations)],
    ),
)
"""

write(joinpath(repo_root, config_path), config_text)
println("Wrote config: ", joinpath(repo_root, config_path))
println(config_text)
```

```julia
cd(repo_root) do
    run(`julia --project=. $run_script $config_path --dry-run`)
end
```

```julia
# Remove `--dry-run` below when you want a real run.
cd(repo_root) do
    run(`julia --project=. $run_script $config_path`)
end
```
